```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = 'plotly_white'
```

```python
data = pd.read_csv('/content/Sample - Superstore.csv' , encoding = 'latin-1')
data.head(3)
```

```python
data.describe()
```

```python
data.info()
```

```python
#Converting Date column

data['Order Date'] = pd.to_datetime(data['Order Date'])
data.head(3)
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
data.head(3)
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day'] = data['Order Date'].dt.day
data.head(3)
```

**Monthly Sales** **Analysis**


```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
sales_by_month
```

```python
plt.figsize = (4,3)
fig = px.line(sales_by_month, x='Order Month', y='Sales', title='Monthly Sales Analysis')
fig.show()
```

**Sales by Category**


```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
sales_by_category
```

```python
fig = px.pie(sales_by_category, values='Sales', names='Category',hole = 0.5 , color_discrete_sequence = px.colors.qualitative.Pastel )
fig.update_traces(textposition ='inside' , textinfo ='percent+label')
fig.update_layout(title_text = 'Sales Analysis by Category' , title_font = dict(size = 24))
fig.show()
```

Sales Analysis by Category

```python
sales_by_sub_category = data.groupby('Sub-Category')['Sales'].sum().reset_index()
sales_by_sub_category
```

```python
fig = px.bar(sales_by_sub_category, x='Sub-Category', y='Sales', title='Sales Analysis by Sub-Category')
fig.show()
```

Monthly Profit Analysis

```python
profit_by_month = data.groupby('Order Month')['Profit'].sum().reset_index()
profit_by_month
```

```python
fig = px.line(profit_by_month, x='Order Month', y='Profit', title='Monthly Profit Analysis')
fig.show()
```

```python
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
profit_by_category
```

```python
fig = px.pie(profit_by_category, values='Profit', names='Category',hole = 0.5 , color_discrete_sequence = px.colors.qualitative.Pastel )
fig.update_traces(textposition ='inside' , textinfo ='percent+label')
fig.show()
```

**Profit by sub-category**

```python
profit_by_sub_category = data.groupby('Sub-Category')['Profit'].sum().reset_index()
profit_by_sub_category
```

```python
fig =px.bar(profit_by_sub_category, x='Sub-Category', y='Profit', title='Profit Analysis by Sub-Category')
fig.show()
```

**Sales and Profit - Customer Segment**

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
color_palette = colors.qualitative.Pastel

fig =go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'], y=sales_profit_by_segment['Sales'], name='Sales', marker_color=color_palette[0]))
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'], y=sales_profit_by_segment['Profit'], name='Profit', marker_color=color_palette[1]))

fig.update_layout(title='Sales and Profit Analysis by Customer Segment', xaxis_title='Segment', yaxis_title='Amount')
fig.show()
```

Sales to Profit Ratio

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment' , 'Sales_to_Profit_Ratio']])
```
